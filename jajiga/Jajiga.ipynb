```python
train_data = pd.read_csv('../data/train.csv', index_col=0).reset_index(drop=True)
test_data = pd.read_csv('../data/test.csv', index_col=0).reset_index(drop=True)
```

```python
df_train_d = train_data.drop(columns=['sleep_description', 'stays_max', 
                                      'additional_feature', 'additional_safety', 
                                      'additional_rule', 'province.name', 
                                      'province.url', 'city.name', 'city.url',
                                      'id'])
df_test_d = test_data.drop(columns=['sleep_description', 'stays_max', 
                                      'additional_feature', 'additional_safety', 
                                      'additional_rule', 'province.name', 
                                      'province.url', 'city.name', 'city.url',
                                      'id'])
```

```python
# drop descriptions
df_train_dd = df_train_d.drop(columns=['title', 'description'])
df_test_dd = df_test_d.drop(columns=['title', 'description'])
```

```python
def get_sleep_arrage_features(x):
    single = 0
    double = 0
    mattress = 0
    for dict_t in ast.literal_eval(x):
        single += dict_t['single']
        double += dict_t['double']
        mattress += dict_t['mattress']
    r_dict = {'single': single, 'double':double, 'mattress':mattress}
    return r_dict
```

```python
df_train_dd['sleep_arrange'] = df_train_dd['sleep_arrange'].map(get_sleep_arrage_features)
df_test_dd['sleep_arrange'] = df_test_dd['sleep_arrange'].map(get_sleep_arrage_features)
```

```python
dummy_sleeparrange_trin = pd.json_normalize(df_train_dd['sleep_arrange'])
dummy_sleeparrange_test = pd.json_normalize(df_test_dd['sleep_arrange'])
```

```python
def get_sleep_arrage_rules(x):
    party = 0
    pet = 0
    smoke = 0
    unmarried = 0
    
    list_t = ast.literal_eval(x)
    if 'party' in list_t:
        party = 1
    if 'pet' in list_t:
        pet = 1
    if 'smoke' in list_t:
        smoke = 1
    if 'unmarried' in list_t:
        unmarried = 1

    r_dict = {'party': party, 'pet':pet, 'smoke':smoke, 'unmarried':unmarried}
    return r_dict
```

```python
df_train_dd['rules'] = df_train_dd['rules'].map(get_sleep_arrage_rules)
df_test_dd['rules'] = df_test_dd['rules'].map(get_sleep_arrage_rules)
```

```python
dummy_rules_trin = pd.json_normalize(df_train_dd['rules'])
dummy_rules_test = pd.json_normalize(df_test_dd['rules'])
```

```python
# drop features
df_train_dd = df_train_dd.drop(columns=['features'])
df_test_dd = df_test_dd.drop(columns=['features'])
```

```python
dummy_types_train = pd.get_dummies(df_train_dd['types'].map(ast.literal_eval).explode())
dummy_types_train = dummy_types_train.groupby(dummy_types_train.index).sum()

dummy_types_test = pd.get_dummies(df_test_dd['types'].map(ast.literal_eval).explode())
dummy_types_test = dummy_types_test.groupby(dummy_types_test.index).sum()
# dummy_types_test
```

```python
dummy_regions_train = pd.get_dummies(df_train_dd['regions'].map(ast.literal_eval).explode())
dummy_regions_train = dummy_regions_train.groupby(dummy_regions_train.index).sum()

dummy_regions_test = pd.get_dummies(df_test_dd['regions'].map(ast.literal_eval).explode())
dummy_regions_test = dummy_regions_test.groupby(dummy_regions_test.index).sum()
# dummy_regions_test
```

```python
dummy_province_trin = pd.get_dummies(df_train_dd['province.id'], dtype='int')
dummy_province_test = pd.get_dummies(df_test_dd['province.id'], dtype='int')
# dummy_province_trin
```

```python
dummy_city_trin = pd.get_dummies(df_train_dd['city.id'], dtype='int')
dummy_city_trin = dummy_city_trin.drop(columns=dummy_city_trin.sum()[(dummy_city_trin.sum()<50)].index)

dummy_city_test = pd.get_dummies(df_test_dd['city.id'], dtype='int')
dummy_city_test = dummy_city_test.drop(columns=dummy_city_test.sum()[(dummy_city_test.sum()<10)].index)
# dummy_city_trin
```

```python
dummy_allocation_trin = pd.get_dummies(df_train_dd['allocation'], dtype='int')
dummy_allocation_test = pd.get_dummies(df_test_dd['allocation'], dtype='int')
# dummy_allocation_trin
```

```python
df_train_dd_f = pd.concat([df_train_dd, dummy_sleeparrange_trin, 
                           dummy_rules_trin, dummy_types_train, 
                           dummy_regions_train, dummy_province_trin, 
                           dummy_city_trin, 
                           dummy_allocation_trin], axis=1)
df_train_dd_f = df_train_dd_f.drop(columns=['sleep_arrange', 'rules', 'types', 
                                            'regions', 'province.id', 'city.id',
                                            'allocation', 'status', 'floors_count'])
# df_train_dd_f['min_price'] = df_train_dd_f['min_price'] / df_train_dd_f['min_price'].max()

df_train_dd_f.columns = df_train_dd_f.columns.astype(str)
df_train_dd_f = df_train_dd_f.astype('float')
```

```python
df_test_dd_f = pd.concat([df_test_dd, dummy_sleeparrange_test, 
                           dummy_rules_test, dummy_types_test, 
                           dummy_regions_test, dummy_province_test, 
                           dummy_city_test, 
                           dummy_allocation_test], axis=1)
df_test_dd_f = df_test_dd_f.drop(columns=['sleep_arrange', 'rules', 'types', 
                                            'regions', 'province.id', 'city.id',
                                            'allocation', 'status', 'floors_count'])

df_test_dd_f.columns = df_test_dd_f.columns.astype(str)
df_test_dd_f = df_test_dd_f.astype('float')
```

```python
X_train, y_train = df_train_dd_f.drop(columns=['min_price']), df_train_dd_f['min_price']
X_train_n = X_train / X_train.max(axis=0)
y_train_n = y_train / df_train_dd_f['min_price'].max()
df_train_dd_f_n = pd.concat([X_train_n, y_train_n],axis=1)

X_test_n = df_test_dd_f / X_train.max(axis=0)
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
پس از پاکسازی داده‌ها و احتمالا حذف بعضی ستون‌ها یا افزودن ویژگی‌های جدید، حال وقت آن است که مطابق سلیقه‌ی خود، مدل مناسب را با استفاده از داده‌های پیش‌پردازش شده آموزش دهید.

```python
def train_model(model, train_data, test_data):
    model.fit(train_data.drop(columns=['min_price']), train_data['min_price'])
    y_pred = model.predict(test_data.drop(columns=['min_price']))
    f1 = r2_score(test_data['min_price'], y_pred)
    return f1
```

```python
random_state = [96, 885, 11445, 62, 3, 7799551, 697]
sum_score = 0
for rr in random_state:
    x_train, x_test = train_test_split(df_train_dd_f_n, test_size=0.2, random_state=rr)
    reg = Ridge(alpha=5)
    r2 = train_model(reg, x_train, x_test)
    sum_score = sum_score + r2
    print(np.round(r2,3)*250)

print(f'average: {sum_score/len(random_state)}')
```

```python
random_state = [96, 885, 11445, 62, 3, 7799551, 697]
sum_score = 0
for rr in random_state:
    x_train, x_test = train_test_split(df_train_dd_f_n, test_size=0.2, random_state=rr)
    reg = XGBRegressor()
    r2 = train_model(reg, x_train, x_test)
    sum_score = sum_score + r2
    print(np.round(r2,3)*250)

print(f'averate: {sum_score/len(random_state)}')
```

```python
random_state = [96, 885, 11445, 62, 3, 7799551, 697]
sum_score = 0
for rr in random_state:
    x_train, x_test = train_test_split(df_train_dd_f_n, test_size=0.2, random_state=rr)
    reg = RandomForestRegressor()
    r2 = train_model(reg, x_train, x_test)
    sum_score = sum_score + r2
    print(np.round(r2,3)*250)

print(f'averate: {sum_score/len(random_state)}')
```
